```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('documents/train1.csv')
df.head(15)
```

```python
df.tail(10)
```

```python
# Variable ----- Description

# Loan_ID -->> Unique Loan ID

# Gender -->> Male/ Female

# Married -->> Applicant married (Y/N)

# Dependents -->> Number of dependents

# Education -->> Applicant Education (Graduate/ Under Graduate)

# Self_Employed -->> Self employed (Y/N)

# ApplicantIncome -->> Applicant income

# CoapplicantIncome -->> Coapplicant income

# LoanAmount -->> Loan amount in thousands of dollarss

# Loan_Amount_Term -->> Term of loan in months

# Credit_History -->> credit history meets guidelines

# Property_Area -->> Urban/ Semi Urban/ Rural

# Loan_Status -->> Loan approved (Y/N)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Loan_Status'].replace('Y',1,inplace=True)
df['Loan_Status'].replace('N',0,inplace=True)
```

```python
df['Dependents'].replace('0',0,inplace=True)
df['Dependents'].replace('1',1,inplace=True)
df['Dependents'].replace('2',2,inplace=True)
df['Dependents'].replace('3+',3,inplace=True)
```

```python
df['Self_Employed'].replace('Yes',1,inplace=True)
df['Self_Employed'].replace('No',0,inplace=True)
```

```python
df.Married=df.Married.map({'Yes':1,'No':0})
```

```python
df.Gender=df.Gender.map({'Male':1,'Female':0})
```

```python
df.Education=df.Education.map({'Graduate':1,'Not Graduate':0})
```

```python
df.Property_Area=df.Property_Area.map({'Urban':2,'Rural':0,'Semiurban':1})
```

```python
df=df.drop(["Loan_ID"],axis=1)
```

```python
colnames=df.columns
colnames
```

# KNN Imputer

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
data_filled = imputer.fit_transform(df)
df1=pd.DataFrame(data=data_filled,columns=colnames)
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
cor_mat=df1.corr()
cor_mat
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.subplots(figsize=(20,15))
sns.heatmap(cor_mat,vmax=1,vmin=-1,annot=True,linewidths=1,cmap='bwr')
```

```python
Loan_Status_counts = df1['Loan_Status'].value_counts()

plt.bar(Loan_Status_counts.index, Loan_Status_counts.values, color=['skyblue', 'pink'])
plt.title('Distribution of Loan Staus')
plt.xlabel('Loan_Status')
plt.ylabel('Count')
plt.xticks(Loan_Status_counts.index, ['yes', 'No'])
plt.show()
```

```python
y=df1['Loan_Status']
X=df1.drop('Loan_Status',axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=40)  
```

# Logistic Regression

```python
#Logistic regression estimates the probability of an event occurring
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
X_test.head(5)
```

```python
lr_prediction = model.predict(X_test)
lr_prediction
```

```python
from sklearn.metrics import confusion_matrix  
cf=confusion_matrix(y_test,lr_prediction) 
cf
```

```python
import seaborn as sns           
sns.heatmap(cf, annot=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,lr_prediction)
accuracy
```

# RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix   

cf=confusion_matrix(y_test,y_pred)
cf
```

```python
import seaborn as sns
sns.heatmap(cf, annot=True)
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
